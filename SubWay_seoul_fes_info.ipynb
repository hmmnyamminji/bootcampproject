```python
import requests
import pandas as pd

API_KEY = "696c51614b67696d3736727678634c"
SERVICE_NAME = "culturalEventInfo"

url = f"http://openapi.seoul.go.kr:8088/{API_KEY}/json/{SERVICE_NAME}/1/100/"

response = requests.get(url, timeout=10)
response.raise_for_status()

data = response.json()
df = pd.DataFrame(data['culturalEventInfo']['row'])

df.replace(0, np.nan, inplace=True)
df.fillna(method='ffill', inplace=True)
df
```
