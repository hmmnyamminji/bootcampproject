```python
import requests
import pandas as pd
import numpy as np

API_KEY = "696c51614b67696d3736727678634c"

url = f"http://openapi.seoul.go.kr:8088/{API_KEY}/json/stadiumScheduleInfo/1/1000/"

response = requests.get(url, timeout=10)
response.raise_for_status()

data = response.json()

df = pd.DataFrame(data['stadiumScheduleInfo']['row'])

print(df.head())
```

```python
df_jamsil = df[df['CODE_TITLE_B'].str.contains('잠실', na=False)].copy()

df_jamsil['SDATE'] = pd.to_datetime(df_jamsil['SDATE'], format='%Y%m%d', errors='coerce')
df_jamsil['EDATE'] = pd.to_datetime(df_jamsil['EDATE'], format='%Y%m%d', errors='coerce')

df_jamsil = df_jamsil[
    (df_jamsil['SDATE'].dt.year >= 2023) & (df_jamsil['SDATE'].dt.year <= 2025)
].copy()

print(df_jamsil.head())
```

```python
columns_to_drop = ['LINK_URL','USE_PAY','CODE_TITLE_A','SCH_CODE_B','SCH_CODE_A','REG_DATE', 'UPD_DATE','USE_AGE','USE_TARGET','SCH_SEQ']
df_jamsil = df_jamsil.drop(columns=columns_to_drop, errors='ignore')
```
